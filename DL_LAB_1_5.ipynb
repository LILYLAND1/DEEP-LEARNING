```python
# Experiment 1: Implementation of a Single Artificial Neuron
import numpy as np

# Define inputs, weights, and bias
inputs = np.array([1, 0])  # Example inputs
weights = np.array([0.8, 0.4])  # Example weights
bias = -0.3  # Example bias

# Activation function: Step function
def step_function(x):
    return 1 if x >= 0 else 0

# Calculate weighted sum (z)
z = np.dot(inputs, weights) + bias

# Apply activation function
output = step_function(z)

print("Input values:", inputs)
print("Weights:", weights)
print("Bias:", bias)
print("Weighted sum (z):", z)
print("Output (activation result):", output)
```

```python
#Experiment 2: Implementation of Single Layer Perceptron (SLP)
import numpy as np

# Define input and output for the AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Inputs
y = np.array([0, 0, 0, 1])  # Outputs

# Initialize weights and bias
weights = np.random.rand(2)  # Random weights
bias = np.random.rand(1)  # Random bias
learning_rate = 0.1  # Learning rate

# Activation function: Step function
def step_function(x):
    return 1 if x >= 0 else 0

# Training the perceptron
epochs = 10  # Number of iterations
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}")
    for i in range(len(X)):
        # Calculate the weighted sum
        z = np.dot(X[i], weights) + bias

        # Apply activation function
        y_pred = step_function(z)

        # Calculate error
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Error: {error}")
    print(f"Updated Weights: {weights}, Updated Bias: {bias}\n")

# Testing the perceptron
print("Final Model Testing:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]}, Predicted Output: {y_pred}")
```

```python
#Experiment 4: Activation Functions (Sigmoid, ReLU, Tanh)
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate a sample dataset
x = np.linspace(-10, 10, 100)  # Input values

# Apply activation functions
sigmoid_output = sigmoid(x)
relu_output = relu(x)
tanh_output = tanh(x)

# Plot the activation functions
plt.figure(figsize=(10, 6))

# Sigmoid
plt.subplot(3, 1, 1)
plt.plot(x, sigmoid_output, label='Sigmoid', color='blue')
plt.title('Sigmoid Activation Function')
plt.grid()
plt.legend()

# ReLU
plt.subplot(3, 1, 2)
plt.plot(x, relu_output, label='ReLU', color='green')
plt.title('ReLU Activation Function')
plt.grid()
plt.legend()

# Tanh
plt.subplot(3, 1, 3)
plt.plot(x, tanh_output, label='Tanh', color='red')
plt.title('Tanh Activation Function')
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
